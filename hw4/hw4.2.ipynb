#### 1. Выкачайте главную страницу сайта https://stihi.ru/. Вытащите из неё ссылку, которая ведет к странице с полным списком рекомендуемых произведений (выбор редакции портала). Она будет лежать в аргументe href у тега, поэтому доставать нужно регуляркой.

Зачем регулярки. С помощью bs ищем сам тэг и выковыриваем href:

```python
import requests
from bs4 import BeautifulSoup

uri = 'https://stihi.ru/'
r = requests.get(url=uri)
text = r.text
soup = BeautifulSoup(text,'html.parser')
post = soup.find('a', {'class': 'topref'})

uri2 = ''.join([uri.strip('/'),post['href']])

print(uri2)
```

#### 2.1. Вытащите со страницы со рекомендуемых произведений соответствия авторов, названий произведений и времени публикации по url, который вы получили в первом пункте с помощью BeautifulSoup или регулярок.

```python
r = requests.get(url=uri2)
text = r.text
soup = BeautifulSoup(text,'html.parser')

authors = []
for author in soup.find_all('a', class_='authorlink'):
    authors.append(author.get_text())
    
creations = []
for creation in soup.find_all('a', class_='poemlink'):
    creations.append(creation.get_text())

dates = []
for date in soup.find_all('small'):
    dates.append(date.get_text().strip('- '))
```

#### 2.2. Сортируйте по времени публикаций и распечатайте.

```python
import pandas as pd

obj_for_df = {
    'authors': authors,
    'creations': creations,
    'dates': dates
}
df = pd.DataFrame(obj_for_df)
df.sort_values(by=['dates'])  # отсортировано по возрастанию, т.к. в задании не сказано, как именно
```
